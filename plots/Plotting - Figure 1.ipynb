```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.charts import Bar
from bokeh.layouts import row
from bokeh.io import output_notebook
from bokeh.charts.attributes import CatAttr
#from bokeh.palettes import Category20_12
```

```python
books = pd.read_pickle("books_additional_columns.pkl")
```

```python
books.head()
```

```python
books["rating"] = books["rating"].map(np.float64)
```

```python
books["year"] = books["year"].map(np.float64)
```

```python
books["num_ratings"] = books["num_ratings"].map(np.float64)
```

```python
books["num_reviews"] = books["num_reviews"].map(np.float64)
```

```python
books = books[books.rating != 0]
```

```python
books.to_pickle("data_figure1.pkl")
```

```python
p = figure(plot_width=400, plot_height=400)

# add a circle renderer with a size, color, and alpha
p.circle(books["year"], books["rating"], size=8, color="navy",alpha=0.5)

# show the results
show(p)
```
