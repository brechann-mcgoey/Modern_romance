```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.charts import Bar
from bokeh.layouts import row
from bokeh.io import output_notebook
from bokeh.charts.attributes import CatAttr
```

#### Reviews and ratings

```python
reviews_ratings = pd.read_pickle("Reviews_ratings.pkl")
```

```python
reviews_ratings.head()
```

```python
reviews_freq = reviews_ratings[reviews_ratings["count"]>10]
```

```python
len(reviews_freq)
```

```python
reviews_freq["impact"]=abs(reviews_freq["coef"])
```

```python
reviews_freq.head()
```

```python
reviews_freq.sort_values("impact",inplace=True,ascending=False)
```

```python
reviews_freq["order"]=list(range(1,3483))
```

```python
index_reviews=[130875,116469,48883,37447,68263,67563,67885,16812,29372,129309]
```

```python
review_subset = reviews_freq.loc[index_reviews]
```

```python
review_subset
```

```python
review_subset.to_pickle("Reviews_ratings_subset.pkl")
```

```python
bar_reviews=Bar(review_subset,values="coef",label=CatAttr(columns=['labels'], sort=False),agg="mean",plot_width=400,
        legend=False,xlabel="Word", ylabel="Coefficient")
show(bar_reviews)
```

#### descriptions and years

```python
description_word_year = pd.read_pickle("descriptions_words_years.pkl")
```

```python
description_word_year.head()
```

```python
key_words=["series","dominant", "bdsm", "vampire", "playboy", 
          "heaven","sexy"]
```

```python
words_years=pd.DataFrame({"year":description_word_year["year"]})
for word in key_words:
    words_years[word]=description_word_year["description_frequency"].map(lambda x: x.get(word))
```

```python
words_years.tail()
```

```python
# select a palette
from bokeh.palettes import Category20_12
# itertools handles the cycling
import itertools 
```

```python
p = figure(plot_width=600, plot_height=300,x_range=(1990, 2016))

##plot_words=words_years("year", 1)
#colors = itertools.cycle(palette) 

for i, word in enumerate(key_words):
    p.line(words_years["year"], words_years[word], line_width=2, color=Category20_12[i],legend=word)

p.legend.location='top_left'    
show(p)
```
