```python
def num_ratings(page):
    try:
        page=BeautifulSoup(page,"lxml")
        ratings=page.find_all("span",class_="value-title")[0].get_text().split()[0].replace(',','')
        ratings=int(ratings)
    except:
        ratings=''
    return ratings
```

```python
books["num_ratings"] = books['page'].apply(num_ratings)
```

```python
def num_reviews(page):
    try:
        page=BeautifulSoup(page,"lxml")
        reviews=page.find_all("span",class_="value-title")[1].get_text().split()[0].replace(',','')
        reviews=int(reviews)
    except:
        reviews=''
    return reviews
```

```python
books["num_reviews"] = books['page'].apply(num_reviews)
```
