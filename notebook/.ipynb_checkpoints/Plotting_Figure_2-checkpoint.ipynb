#### descriptions and years

```python
description_word_year = pd.read_pickle("descriptions_words_years.pkl")
```

```python
description_word_year.head()
```

```python
key_words=["series","dominant", "bdsm", "vampire", "playboy", 
          "heaven","sexy"]
```

```python
words_years=pd.DataFrame({"year":description_word_year["year"]})
for word in key_words:
    words_years[word]=description_word_year["description_frequency"].map(lambda x: x.get(word))
```

```python
words_years.tail()
```

```python
# select a palette
from bokeh.palettes import Category20_12
# itertools handles the cycling
import itertools 
```

```python
key_words=['princess','queen','king','royalty','royal','crown', 'castle','military', 'navy', 'army', 'soldier',
          'vampire','werewolf','ghost','paranormal','supernatural','psychic','witch', 'christmas', 'thanksgiving', 
           'valentine', 'halloween', 'new year','bdsm','dominant','submissive','quest','warrior','voyage',
           'journey','adventure']
```

```python
words_years2=pd.DataFrame({"mo_yr":reviews_words_years["date_formatted"]})
for word in key_words:
    words_years2[word]=reviews_words_years["review_date_frequency"].map(lambda x: x.get(word))
```

```python
words_years2['review_count'] = reviews_words_years['counts']
```

```python
words_years2.drop([22,77,23],axis=0,inplace=True)
```

```python
words_years2.tail()
```

```python
plot_data=words_years2[['dts','Royalty','Military','Supernatural','Holiday','Bdsm','Adventure']]
```

```python
plot_data.drop([100],axis=0,inplace=True)
```

```python
plot_data.tail()
```

```python
plot_data.to_pickle("themes_time.pkl")
```

```python
words_years2["dts"]=pd.to_datetime(words_years2["mo_yr"])
```

```python
words_years2.sort_values('dts',inplace=True)
```

```python
royalty=['princess','queen','king','royalty','royal','crown', 'castle']
military=['military', 'navy', 'army', 'soldier']
supernatural=['vampire','werewolf','ghost','paranormal','supernatural','psychic','witch']
holiday=['christmas', 'thanksgiving', 'valentine', 'halloween']
bdsm=['bdsm','dominant','submissive']
adventure=['quest','warrior','voyage','journey','adventure']
```

```python
words_years2['Royalty']=words_years2[royalty].sum(axis=1)/words_years2['review_count']
```

```python
words_years2['Military']=words_years2[military].sum(axis=1)/words_years2['review_count']
words_years2['Supernatural']=words_years2[supernatural].sum(axis=1)/words_years2['review_count']
words_years2['Holiday']=words_years2[holiday].sum(axis=1)/words_years2['review_count']
words_years2['Bdsm']=words_years2[bdsm].sum(axis=1)/words_years2['review_count']
words_years2['Adventure']=words_years2[adventure].sum(axis=1)/words_years2['review_count']
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.models import DatetimeTickFormatter

# Set up data

x = plot_data['dts']
y = plot_data['Royalty']

source = ColumnDataSource(data=dict(x=x, y=y))


# Set up plot
plot = figure(plot_height=400, plot_width=800, title="Themes",tools="crosshair,pan,reset,save,wheel_zoom")

plot.line(x='x', y='y', source=source, line_width=3, line_alpha=0.6)

plot.xaxis.formatter=DatetimeTickFormatter(
        hours=["%B %Y"],
        days=["%B %Y"],
        months=["%B %Y"],
        years=["%B %Y"],
    )
plot.xaxis.major_label_orientation = np.pi/4
output_notebook()
show(plot)

```
